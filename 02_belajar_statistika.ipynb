# 2. Klasifikasi Data Dalam Statistika

## Tipe Data

Dalam statistika, terdapat dua ilmu data, yaitu:

- **Data Kualitatif (Qualitative)**: Data kualitatif berasosiasi dengan data-data non-numerik, seperti attribute dan label.

- **Data Kuantitatif (Quantitative)**: Data kuantitatif berasosiasi dengan data-data numerik yang dihasilkan melalui proses penghitungan atau pengukuran.

## Contoh Tipe Data Kualitatif Dan Kuantitatif

```python
import pandas as pd
```

```python
data = {
    "Menu": ["Nasi Goreng", "Nasi Pecel", "Nasi Timbel"],
    "Harga": [25_000, 30_000, 35_000]
}

# Make the DataFrame object
df = pd.DataFrame(data)
df
```

Pada kolom Menu, terdapat tiga isi data makanan: Nasi Goreng, Nasi Pecel, dan Nasi Timbel. Ketiga data ini adalah data non-numerik, dan bisa kita masukan sebagai tipe data kualitatif. Selanjutnya adalah kolom Harga, pada kolom harga terdapat tiga harga makanan: Rp. 25,000, Rp. 30,000, dan Rp. 35,000. Ketiga data harga makanan ini adalah data numerik, dan bisa kita masukan sebagai tipe data kuantitatif.

```python
tabel_harga = df.style.format({"Harga": "Rp. {:,}"})

tabel_harga.set_caption("Tabel Harga Makanan")

tabel_harga.hide()
```

## Skala Pengukuran (Level Of Measurements)

Terdapat 4 skala pengukuran (level of measurements) dalam ilmu statistika: nominal, ordinal, interval, dan rasio.

### Skala Pengukuran Nominal

- Berasosiasi dengan tipe data kualitatif
- Berfokus pada pengelompokan atau pengkategorian data berdasarkan nama, label, atau kualitas
- Tidak dapat dikenakan operasi matematika
- **Contoh**:
  - Merek kendaraan bermotor
  - Kota kelahiran
  - Nama siswa, nama bulan, nama hari
  - Nomor kendaraan bermotor, nomor induk siswa

## Skala Pengukuran Ordinal

- Berasosiasi dengan tipe data kualitatif
- Data dapat dikelompokan
- Data dapat disusun berdasarkan urutan, peringkat, atau rangking
- Tidak dapat dikenakan operasi matematika
- **Contoh**:
  - Tingkat kepuasaan pelanggan: sangat puas, puas, cukup, kecewa, dan sangat kecewa
  - Temperatur udara: panas, sejuk, dan dingin
  - Peringkat siswa di kelas

## Skala Pengukuran Interval

- Berasosiasi dengan tipe data kuantitatif
- Data dapat dikelompokan
- Data dapat disusun berdasarkan urutan, peringkat, atau rangking
- Dapat dikenakan operasi aritmatika untuk menghitung selisih nilai
- Nilai numerik mempresentasikan posisi pada suatu skala ukur tertentu
- Tidak memiliki nol yang absolut (has no inherent zero)
- Tidak dapat dikenakan operasi matematika terkait perkalian ataupun pembagian
- **Contoh**:
  - Temperatur udara 32 derajat celcius
  - Tahun 2020

## Skala Pengukuran Rasio (Ratio)

- Berasosiasi dengan tipe data kuantitatif
- Data dapat dikelompokan
- Data dapat disusun berdasarkan urutan, peringkat, atau rangking
- Dapat dikenakan operasi aritmatika untuk menghitung selisih nilai
- Memiliki nilai nol yang absolut (has inherent zero)
- Dapat dikenakan operasi aritmatika terkait perkalian
- **Contoh**:
  - Usia anak 7 tahun
  - Berat badan 65 kg
  - Harga nasi goreng Rp. 25,000
